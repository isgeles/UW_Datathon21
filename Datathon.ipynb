## Data Visualization

## Model selection

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
# selecting columns
data = df[selected_columns]
targets = df[['Diabetes_binary']]

# splitting the data in training and validation set
X_train, X_val, y_train, y_val = train_test_split(data, targets, test_size=0.20, random_state=123)
y_train = y_train.values.ravel()
y_val = y_val.values.ravel()


X_train.shape, X_val.shape, y_train.shape, y_val.shape
```

### SVM

```python
clf = SVC()
clf.fit(X_train, y_train)
print("Accuracy achieved with the validation set: ", clf.score(X_val, y_val))
```

### Decision Tree 

```python
clf = DecisionTreeClassifier()
clf.fit(X_train, y_train)
print("Accuracy achieved with the validation set: ", clf.score(X_val, y_val))
```

### Random Forest

```python
clf = DecisionTreeClassifier()
clf.fit(X_train, y_train)
print("Accuracy achieved with the validation set: ", clf.score(X_val, y_val))
```

### Neural Network

```python
import tensorflow as tf
from tensorflow import keras
from keras.models import Sequential 
from keras.layers import * 


# Creating a simple sequential model
model = tf.keras.Sequential()
model.add(tf.keras.Input(shape=(len(selected_columns),))) 
model.add(tf.keras.layers.Dense(100, activation='relu'))
model.add(tf.keras.layers.Dense(100, activation='relu'))
model.add(tf.keras.layers.Dense(100, activation='relu'))
model.add(tf.keras.layers.Dense(1, activation='sigmoid'))

model.summary()

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])


history = model.fit(X_train, y_train, epochs=20, batch_size=128, verbose=1)

# Evaluate test accuracy
test_loss, test_acc = model.evaluate(X_val, y_val)

print("\nTesting with the validation set \nloss: {:.3f} \t accuracy {:.3f}".format(test_loss, test_acc))
```

```python
# plotting training progress

fig = plt.figure(figsize=(12,5))
plt.subplot(1, 2, 1)
plt.plot(history.history['loss'])
plt.ylabel('train loss')
plt.xlabel('epochs')

plt.subplot(1, 2, 2)
plt.plot(history.history['accuracy'])
plt.ylabel('train accuracy')
plt.xlabel('epochs')
plt.suptitle("Neural Network Training Progress")
plt.tight_layout()
plt.show()
```

## Conclusion

With our best performing models (SVM or fully connected neural network) we are able to predict with over **74% accuracy**
